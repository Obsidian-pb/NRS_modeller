# Расчет глубин тушения ручных пожарных стволов для различных параметров

```python
import math

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]
```

```python
def get_h(q, i, acc=1):
  '''
  Функция расчета площади, глубины и фронта тушения
  для заданных производительности ствола и интенсивности подачи ОТВ.

  `q` - Производительность ствола

  `i` - требуемая интенсивность подачи ОТВ

  `acc` = 1
    Округление до `acc` знаков

  '''
  s = q / i
  h = math.sqrt( 2 * s / math.pi )
  f = 2 * round(h, acc)
  return round(s, acc), round(h, acc), round(f, acc)

# Пример расчета
s, h, f = get_h(3.7 , 0.03)
print('Площадь тушения:', s)
print('Глубина тушения:', h)
print('Фронт тушения:  ', f)
```

```python
fig, ax = plt.subplots(1,1, tight_layout=True)  #, dpi=300)

ia = [q*0.01 for q in range(1, 30)]

# Для производительности 3,7
ha = []
q = 3.7
for i in ia:
  s, h, f = get_h(q, i, acc=4)
  ha.append(h)
ax.plot(ia, ha, label='$q_{ст}$ = 3,7 л/с (РС-50)')

# Для производительности 7,4
ha = []
q = 7.4
for i in ia:
  s, h, f = get_h(q , i, acc=4)
  ha.append(h)
ax.plot(ia, ha, label='$q_{ст}$ = 7,4 л/с (РС-70)')

# Для производительности 8
ha = []
q = 8
for i in ia:
  s, h, f = get_h(q , i, acc=4)
  ha.append(h)
ax.plot(ia, ha, label='$q_{ст}$ = 8 л/с (УРСК-50)')

# Для производительности 15
ha = []
q = 15
for i in ia:
  s, h, f = get_h(q , i, acc=4)
  ha.append(h)
ax.plot(ia, ha, label='$q_{ст}$ = 15 л/с (УРСК-70)')


ax.set_xlabel('Интенсивность подачи воды, $I_{тр}$, $л/с*м^2$')
ax.set_ylabel('Глубина тушения, м')
ax.legend()

ax.grid('on', linestyle=":")
plt.savefig('1.png', dpi=300)
plt.show()
```
