## Расчет простой насосно-рукавной системы

![модель НРС](imgs/nrs_simple_1.jpg)

Схема НРС состоит из:
* 1 источника в виде насоса ПА
* 1 рабочей рукавной линии из 10 рукавов диаметром 51 мм
* 1 ствола Б

```python
# Создаем элементы НРС
# Насос ПА
pump    = Element('Насос',                  # Название
                  e_type = EType.PUMP,      # Тип элемента: НАСОС
                  H_add = 40)               # Стартовый напор на насосе = 40 м                     
# Рабочая рукавная линия
hoseW1  = Element('Рабочая рукавная линия',       # Название
                  e_type = EType.CONNECTOR,         # Тип элемента: рукавная линия
                  s = NRS_Data.ss["51"],            # Гидравлическое сопротивление 1 рукава диаметром 51мм
                  n = 10)                           # Кол-во рукавов=10

nozzle1 = Element('Ствол 1',                             # Название
                  e_type = EType.NOZZLE,                 # Тип элемента: СТВОЛ
                  p = NRS_Revision.calc_p(3.7, 40),      # Проводимость ствола
                  q_out = q_out_nozzle)                  # Для ствола указываем метод расчета производительности

# Соединяем элементы друг с другом Насос -> Рукавная линия -> Ствол
pump.append(hoseW1).append(nozzle1)

# Строим модель, начиная с элемента насоса
model = NRS_Model('Простая модель от одного насоса к одному стволу')
model = model.build(pump)

# Непосредственно расчет модели:
model.calc(accuracy=0.05)
print('Производительность ствола:', round(model.summaryQ(), 2), 'л/с')
```

Проверим результат.

Производительность ствола определяется по формуле:

$$
Q_{ств} = p * \sqrt{H_{ств}}
$$

где, 
* $Q_{ств}$ - производительность ствола
* $p$ - проводимость насадка
* $H_{ств}$ - напор перед стволом

```python
print('Напор на насосе:               ', pump.H_out, 'м')
print('Потеря напора в рукавной линии:', hoseW1.h, 'м')
print('Напор на стволе               :', nozzle1.H_in, 'м')
print('='*80)

q_out_test = NRS_Revision.calc_p(3.7, 40) * pow(nozzle1.H_in, 0.5)
assert q_out_test == model.summaryQ()
print('Рассчитанная производительность', model.summaryQ(), 'л/с')
print('Проверочная производительность ', q_out_test, 'л/с')
print('РАСЧЕТ ВЕРЕН')
```

## Расчет насосно-рукавной системы с 1 стволом А и одним стволом Б

![модель НРС](imgs/nrs_2.jpg)

Схема НРС состоит из:
* 1 источника в виде насоса ПА
* 1 магистральной рукавной линии из 5 рукавов 77мм
* 1 рабочей рукавной линии из 10 рукавов диаметром 51 мм
* 1 ствола Б
* 1 рабочей рукавной линии из 5 рукавов диаметром 77 мм
* 1 ствола А

```python
# Создаем элементы НРС
# Насос ПА
pump    = Element('Насос',                               # Название
                  e_type = EType.PUMP,                   # Тип элемента: НАСОС
                  H_add = 60)                            # Стартовый напор на насосе = 60 м
# Магистральная рукавная линия
hoseM1  = Element('Магистральная рукавная линия',        # Название
                  e_type = EType.CONNECTOR,              # Тип элемента: рукавная линия
                  s = NRS_Data.ss["77"],                 # Гидравлическое сопротивление 1 рукава диаметром 51мм
                  n = 5)                                 # Кол-во рукавов=5
# Разветвление
splitter_1 = Element('Р1',                               # Название
                     EType.CONNECTOR,                    # Тип элемента
                     ro=3)                               # количество выходных патрубков

# Рабочая рукавная линия 1
hoseW1  = Element('Рабочая рукавная линия 1',            # Название
                  e_type = EType.CONNECTOR,              # Тип элемента
                  s = NRS_Data.ss["51"],                 # Гидравлическое сопротивление 1 рукава диаметром 51мм
                  n = 10)                                # Кол-во рукавов=10
nozzle1 = Element('Ствол 1',                             # Название
                  e_type = EType.NOZZLE,                 # Тип элемента: СТВОЛ
                  p = NRS_Revision.calc_p(3.7, 40),      # Проводимость ствола
                  q_out = q_out_nozzle)                  # Для ствола указываем метод расчета производительности

# Рабочая рукавная линия 2
hoseW2  = Element('Рабочая рукавная линия 2',            # Название
                  e_type = EType.CONNECTOR,              # Тип элемента
                  s = NRS_Data.ss["77"],                 # Гидравлическое сопротивление 1 рукава диаметром 51мм
                  n = 5)                                 # Кол-во рукавов=5
nozzle2 = Element('Ствол 2',                             # Название
                  e_type = EType.NOZZLE,                 # Тип элемента: СТВОЛ
                  p = NRS_Revision.calc_p(7.4, 40),      # Проводимость ствола
                  q_out = q_out_nozzle)                  # Для ствола указываем метод расчета производительности

# Соединяем элементы друг с другом Насос -> Рукавная линия -> Ствол
pump.append(hoseM1).append(splitter_1)
splitter_1.append(hoseW1).append(nozzle1)
splitter_1.append(hoseW2).append(nozzle2)

# Строим модель, начиная с элемента насоса
model = NRS_Model('Простая модель от одного насоса к одному стволу')
model = model.build(pump)

# Непосредственно расчет модели:
model.calc(accuracy=0.05)
print('Производительность системы:  ', round(model.summaryQ(), 2), 'л/с')
print('Производительность ствола Б: ', round(nozzle1.get_q_out(), 2), 'л/с')
print('Производительность ствола А: ', round(nozzle2.get_q_out(), 2), 'л/с')
```

```python
print('Потеря напора в магистральной рукавной линии: ', hoseM1.h, 'м')
print('Потеря напора в 1 рабочей рукавной линии:     ', hoseW1.h, 'м')
print('Потеря напора во 2 рабочей рукавной линии:    ', hoseW2.h, 'м')
```
